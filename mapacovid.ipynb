```python
#importação das bibliotecas 
import folium
import pandas as pd

#Leitura das bases de dados necessárias
ler = pd.read_excel('cidades1.xlsx')
tipo = pd.read_excel('tipos.xlsx')

#Geração do mapa que será exibido na tela quando for aberto
mapa = folium.Map(
    location=[-10.4387, -39.3307],
    zoom_start=6,
    tiles = "CartoDB dark_matter"
)


#Criação e designer dos balões de exibição junto com o conjunto de dados que será exibido em cada balão 
for index, linha in ler.iterrows():
    folium.Marker([linha ["LAT"], linha["LONG"]],
    popup= '<b>Localidade:</b> {}<br><b>Confirmados:</b> {}<br><b>Suspeitos:</b> <br>{}<br><b>Monitorados:</b> {}<br><b>Descartados:</b> {}<br><b>OBITOS:<br></b> {}<br>'
            .format(linha["LOCALIDADE"],linha["CONFIRMADOS"], linha["SUSPEITOS"], linha["MONITORADOS"], linha["DESCARTADOS"], linha["OBITOS"]),
    icon=folium.Icon(color=tipo["COR_EXTERNA"][0],
                    icon_color=tipo["COR_INTERNA"][0],
                    icon=tipo["ICONE"][0],
                    prefix=tipo["PREFIX"][0])).add_to(mapa)
#Salvar arquivo final em html para visualização em navegadores da web
mapa.save('covid19.html')
mapa
```
